```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.isna().sum()
```

```python
titanic['age'].fillna(titanic['age'].mean(), inplace=True)
```

```python
titanic['embark_town'].fillna(titanic['embark_town'].mode()[0], inplace=True)
```

```python
sns.countplot(x='embark_town',hue='survived', data=titanic)
plt.title('Overall By Town')
plt.show()
```

```python
titanic.columns
```

```python
sns.countplot(x='who',hue='survived', data=titanic)
plt.title('Overall By who')
plt.show()
```

```python
sns.scatterplot(x='age',y='fare',hue='survived', data=titanic)
plt.title('Age vs Fare')
plt.show()
```

```python
male_survived = titanic[(titanic['sex']=='male') & (titanic['survived']==1)]
sns.violinplot(x='pclass',y='fare', data=male_survived,palette='Set2')
plt.title('Gender dist')
plt.show()
```

```python
diamond = pd.read_csv('/content/diamonds.csv')
```

```python
diamond.head()
```

```python
diamond = diamond.drop('Unnamed: 0',axis=1)
diamond.head()
```

```python
diamond.info()
```

```python
diamond.isna().sum()
```

```python
diamond.shape
```

```python
diamond.dropna(inplace=True)
```

```python
sns.scatterplot(x='carat',y='price',data=diamond)
plt.title('Carat vs Price')
plt.show()
```

```python
diamond['z'] = pd.to_numeric(diamond['z'], errors='coerce')
diamond.dropna(subset=['z'], inplace=True)
numeric_cols = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

corr = diamond[numeric_cols].corr()

plt.figure(figsize=(10, 7))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Diamonds Numeric Features')
plt.show()
```

```python
diamond.drop(['x','y','z'],inplace=True, errors='ignore')
```

```python
price_by_color =  diamond.groupby('color')['price'].mean()
price_by_color = pd.DataFrame(price_by_color)
price_by_color
```

```python
sns.barplot(x='color',y='price',data=price_by_color)
plt.title('Overall By Color')
plt.show()
```
